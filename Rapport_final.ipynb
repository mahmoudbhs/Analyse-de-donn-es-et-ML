##### One-Hot Enconding:
les colonnes qui ont été encodées (de categorie texte) sont :  
Card_Type  
Transaction_Type  
Country  
Device_Type  

Cette méthode consiste à transformer chaque catégorie en une variable binaire.  
Elle est adaptée aux variables qualitatives nominales qui ne peuvent pas etre utilisées directement par un modèle de machine learning, car elle n’introduit pas de relation d’ordre artificielle entre les catégories. 

```python
data = pd.get_dummies(
    data,
    columns=["Merchant_Category", "Card_Type", "Transaction_Type", "Country", "Device_Type"],
    drop_first=True
)

# Securite: XGBoost plante si noms de colonnes dupliques
if not data.columns.is_unique:
    data = data.loc[:, ~data.columns.duplicated()].copy()
```

##### Normalisation:

les colonnes consérnees;
Amount, Distance_From_Home, Hour_Of_Day

-La normalisation ou la standardisation des données est importante car certains modèles de machine learning sont sensibles à l’échelle des variables.  Sans mise à l’échelle, les variables ayant de grandes valeurs numériques peuvent influencer excessivement le modèle par rapport aux autres. Par exemple amount(milliers) et hour_of_day(entre 0-23) sans standardisation, le modèle donnerait beaucoup plus de poids à Amount, même si ce n’est pas forcément la variable la plus importante.


```python
from sklearn.preprocessing import StandardScaler

# Colonnes numeriques a standardiser
num_cols = ["Amount", "Distance_From_Home", "Hour_of_Day"]
missing = [c for c in num_cols if c not in data.columns]
if missing:
    raise KeyError(f"Colonnes manquantes pour la standardisation: {missing}")

scaler = StandardScaler()
data[num_cols] = scaler.fit_transform(data[num_cols])

# Verification rapide
print(data[num_cols].head())
```

##### Transaction_Date:
Date/heure d’une transaction ne pas etre utiliser telle quelle, on doit les transformer en caractéristiques numériques utiles.

```python
data['Transaction_Date'] = pd.to_datetime(data['Transaction_Date'], errors='coerce')
data["Transaction_DayOfWeek"] = data["Transaction_Date"].dt.dayofweek
data["Transaction_Hour"] = data["Transaction_Date"].dt.hour
data["Transaction_Month"] = data["Transaction_Date"].dt.month
data.drop("Transaction_Date", axis=1, inplace=True)
data.dtypes
```

```python
import numpy as np
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import precision_recall_curve, average_precision_score, classification_report, confusion_matrix
import matplotlib.pyplot as plt

# 1) IMPORTANT : X doit être 100% numérique
# Si X est un DataFrame, convertis en numpy
X_np = X.to_numpy(dtype=np.float32) if hasattr(X, "to_numpy") else np.asarray(X, dtype=np.float32)
y_np = y.to_numpy(dtype=np.int32).ravel() if hasattr(y, "to_numpy") else np.asarray(y, dtype=np.int32).ravel()

X_train, X_test, y_train, y_test = train_test_split(
    X_np, y_np, test_size=0.2, stratify=y_np, random_state=42
)

# Poids pour classe positive (fraude)
n_pos = (y_train == 1).sum()
n_neg = (y_train == 0).sum()
scale_pos_weight = n_neg / max(n_pos, 1)

xgb = XGBClassifier(
    n_estimators=500,
    max_depth=6,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    objective="binary:logistic",
    eval_metric="aucpr",
    scale_pos_weight=scale_pos_weight,
    random_state=42,
    n_jobs=-1
)

xgb.fit(X_train, y_train)

y_proba = xgb.predict_proba(X_test)[:, 1]

# Seuil meilleur F1
precisions, recalls, thresholds = precision_recall_curve(y_test, y_proba)
f1_scores = 2 * (precisions * recalls) / (precisions + recalls + 1e-12)
best_f1_idx = np.argmax(f1_scores[:-1])
thr_f1 = thresholds[best_f1_idx]

print("=== Seuil F1 optimal ===")
print(f"Seuil:     {thr_f1:.4f}")
print(f"Precision: {precisions[best_f1_idx]:.4f}")
print(f"Recall:    {recalls[best_f1_idx]:.4f}")
print(f"F1:        {f1_scores[best_f1_idx]:.4f}")

y_pred_f1 = (y_proba >= thr_f1).astype(int)
print("\nClassification report (seuil F1)")
print(classification_report(y_test, y_pred_f1, digits=4))
print(confusion_matrix(y_test, y_pred_f1))

ap = average_precision_score(y_test, y_proba)
print(f"\nPR-AUC (Average Precision): {ap:.4f}")

plt.figure(figsize=(6,4))
plt.plot(recalls, precisions, label=f"PR curve (AP={ap:.4f})")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.title("Precision-Recall Curve (XGBoost)")
plt.grid(True)
plt.legend()
plt.show()
```
